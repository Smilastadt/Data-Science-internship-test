```python
import re
def replace_incorrect_names(df):
    mountain_names = df['label'].tolist()

    for idx, row in df.iterrows():
        text = row['text']
        label = row['label']

        # Find all possible mountain names in the text
        possible_names = [name for name in mountain_names if re.search(rf'\b{name}\b', text, flags=re.IGNORECASE)]

        # If the label is not found in the possible names, replace with the correct label
        if label.lower() not in [name.lower() for name in possible_names]:
            df.at[idx, 'text'] = re.sub(r'\b' + '|'.join(possible_names) + r'\b', label, text, flags=re.IGNORECASE)

    return df
```
